```python
import pandas as pd
```

Load Data

```python
df = pd.read_csv('broken_sales_data.csv')
```

```python
print(df)
```

Checking Missing Values

```python
print(df.isnull().sum())
```

Replacing Nulls

```python
median_sales = df[df["Category"] == "Electronics"]["Sales"].median()

df["Sales"] = df["Sales"].fillna(median_sales)
```

```python
df.loc[df["Product"] == "Chair" , "Quantity"] = 1
```

Detecting Outlier

```python
outlier = df[df["Quantity"] > 100]
print(outlier)
```

Removing Outlier

```python
mean_quantity = df[df["Category"] == "Office"]["Quantity"].mean()
df.loc[df["Quantity"] > 100] = mean_quantity
```

```python
print(df.info())
```

Cleaned Data

```python
print(df)
```

```python
import pandas as pd
import plotly.express as px

# Group by Category
summary = df.groupby("Category")["Sales"].sum().reset_index()

fig = px.bar(
    summary,
    x="Category",
    y="Sales",
    title="Total Sales by Category (Cleaned Data)",
)

# Show plot
fig.show()
```

```python
df.to_csv('cleaned_sales_data.csv', index ='False')
```
